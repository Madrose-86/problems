# data cleaning

```python
telecom_data = telecom.copy()
```

```python
telecom_data.TotalCharges = pd.to_numeric(telecom_data.TotalCharges, errors = 'coerce')
#error = coerce converts missing values to NaN and to find missing values we use next line
telecom_data.isnull().sum()
```

```python
telecom_data.loc[telecom_data['TotalCharges'].isnull() == True]
#it is to identify missing values
```

# missing values treatment

```python
telecom_data.dropna(how='any',inplace = True)
```

```python
print(telecom_data['tenure'].max())
```

```python
# Assuming 'tenure' is the name of the column containing the tenure values
# You can create tenure groups based on specific ranges
bins = [0, 12, 24, 36, 48, 60, 72]
labels = ['1-12', '13-24', '25-36', '37-48', '49-60', '61-72']

# Create a new column 'tenure_group' based on the tenure ranges
telecom_data['tenure_group'] = pd.cut(telecom_data['tenure'], bins=bins, labels=labels)

# Now you can use value_counts() to count the occurrences of each 'tenure_group'
tenure_group_counts = telecom_data['tenure_group'].value_counts()

print(tenure_group_counts)
```

```python
telecom_data.drop(columns=['customerID','tenure'],axis =1, inplace = True)
#here axis=1 to drop columns.
telecom_data.head()
```

# univariate analysis

```python
for i, predictor in enumerate(telecom_data.drop(columns =['Churn','TotalCharges','MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data= telecom_data, x= predictor, hue ='Churn')
```

```python
telecom_data['Churn'] =np.where(telecom_data.Churn =='Yes',1,0)
telecom_data.head()
```

```python
telecom_data_dummies = pd.get_dummies(telecom_data)
telecom_data_dummies.head()
```

```python
sns.lmplot(data = telecom_data_dummies, x= 'MonthlyCharges', y = 'TotalCharges', fit_reg = False)
```

```python
Mth = sns.kdeplot(telecom_data_dummies.MonthlyCharges[(telecom_data_dummies["Churn"] == 0)],color = "Red", shade= True)
Mth = sns.kdeplot(telecom_data_dummies.MonthlyCharges[(telecom_data_dummies["Churn"]==1)], color ="blue", shade = True)
Mth.legend(["No Churn","Churn"], loc ='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly Charges over churn')
```

```python
Tot = sns.kdeplot(telecom_data_dummies.TotalCharges[(telecom_data_dummies["Churn"] == 0)],color = "Red", shade= True)
Tot = sns.kdeplot(telecom_data_dummies.TotalCharges[(telecom_data_dummies["Churn"]==1)], color ="blue", shade = True)
Tot.legend(["No Churn","Churn"], loc ='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total Charges over churn')
```

```python
plt.figure(figsize =(12,12))
plt.xticks(fontsize=14)
plt.yticks(fontsize =14)
sns.heatmap(telecom_data_dummies.corr(), cmap= "Paired")
```
